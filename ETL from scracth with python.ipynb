# **EXTRACT DATA**

## **Import Libraries**




```python
import requests
from bs4 import BeautifulSoup
```

## **Get Html Content with Function**

```python
pages = 4
```

```python
def scraping_web(page):
  url = f"https://webscraper.io/test-sites/e-commerce/static/computers/tablets?page={page}"
  return requests.get(url).text
```

```python
html_content = ''

for page in range(pages):
  temp = scraping_web(page+1)
  html_content += temp
```

```python
html_content
```

## **Parse** **Html** **Content**

```python
parsed_html_content = BeautifulSoup(html_content, 'html.parser')
```

```python
type(parsed_html_content)
```

```python
parsed_html_content
```

## **Get Title**

```python
titles = parsed_html_content.find_all('a', class_= 'title')
```

```python
titles
```

```python
list_title = []

for title in titles:
  list_title.append(title.text)
```

```python
list_title
```

## **Get Price**

```python
prices = parsed_html_content.find_all('h4', class_= 'pull-right price')
```

```python
prices
```

```python
list_price = []

for price in prices:
  list_price.append(price.text)
```

```python
list_price
```

## **Get Description**

```python
descriptions = parsed_html_content.find_all('p', class_= 'description')
```

```python
descriptions
```

```python
list_description = []

for description in descriptions:
  list_description.append(description.text)
```

```python
list_description
```

## **Get Review**

```python
reviews = parsed_html_content.find_all('p', class_='pull-right')
```

```python
reviews
```

```python
list_review = []

for review in reviews:
  list_review.append(review.text)
```

```python
list_review
```

## **Create Dataframe**



```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'title': list_title,
    'price': list_price,
    'description': list_description,
    'jumlah_review': list_review
})
```

```python
df.shape
```

```python
df.head()
```

# **TRANSFORM DATA**

## **Cek Data Type & Data Duplication**

## **Clean Price Column**

```python
df['price'] = df['price'].str.replace('$', '')
```

```python
df.head()
```

## **Convert Price Column to Float**

```python
df['price'] = pd.to_numeric(df['price'])
```

```python
df.head()
```

```python
df.info()
```

## **Clean Jumlah Review Column**

```python
df['jumlah_review'] = df['jumlah_review'].str.replace('reviews', '')
```

```python
df.head()
```

## **Convert Jumlah Review Column to Integer**

```python
df['jumlah_review'] = pd.to_numeric(df['jumlah_review'])
```

```python
df.info()
```

## **Construct Id Column**

```python
edited_price = df['price'].apply(lambda price: str(price))
```

```python
edited_price
```

```python
df.head()
```

```python
edited_title = df['title'].apply(lambda title: title[-6:].replace(' ', '_'))
```

```python
df['id'] = edited_title + '-' + edited_price
```

## **Load Data to DB Using SQL Wrapped by Python**

### **Create Table If Not Exists**

```python
query = """
  CREATE TABLE IF NOT EXISTS tablet_fakri_homework_ETL (
    id VARCHAR(255) PRIMARY KEY NOT NULL,
    title VARCHAR(255),
    price DECIMAL(10,2),
    description VARCHAR(255),
    jumlah_review INT(20)
  )
"""
```

### **Convert Pandas Series to List**

```python
df.head()
```

```python
list_tablet_dict = []

for i in range(len(df)):
  temp_dict = {}
  temp_dict["id_column"] = df.loc[i:i, 'id'].tolist()[0]
  temp_dict["title_column"] = df.loc[i:i, 'title'].tolist()[0]
  temp_dict["price_column"] = df.loc[i:i, 'price'].tolist()[0]
  temp_dict["description_column"] = df.loc[i:i, 'description'].tolist()[0]
  temp_dict["jumlah_review_column"] = df.loc[i:i, 'jumlah_review'].tolist()[0]

  list_tablet_dict.append(temp_dict)
```

```python
list_tablet_dict
```

### **Default Insert Into**

```python
query = """
  INSERT INTO digitalskola.tablet_fakri_homework_ETL (id, title, price, description, jumlah_review)
  VALUES (%s, %s, %s, %s, %s)
"""
```

### **Execute Insert Query**

```python
for tablet in list_tablet_dict:
  data = (tablet["id_column"], tablet["title_column"], tablet["price_column"], tablet["description_column"], tablet["jumlah_review_column"])
  conn.execute(query, data)
```
